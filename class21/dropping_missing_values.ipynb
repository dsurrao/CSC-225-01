## Dropping missing data
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html?highlight=dropna#pandas.DataFrame.dropna

```python
import pandas as pd
import numpy as np
```

```python
df_counties = pd.read_csv('data/live/us-counties.csv', \
                               dtype={'fips': 'string'})
df_counties.head()
```

### Drop unwanted columns

```python
# counties that do not have fips codes
df_counties = df_counties[['state', 'fips']]
df_counties[df_counties['fips'].isna()].head()
```

```python
# drop counties that have null values
#df_counties.dropna(inplace=True)

# should now be empty
df_counties[df_counties['fips'].isna()].head()
```

```python
# create a new column, this will fail if empty values are not dropped!
df_counties['fips_state'] = df_counties['fips'].apply(lambda x: x[:2])
```
