### Map county data

```python
# rename fips
df_states.rename(columns={"fips": "fips_state"}, inplace=True)
df_states.head()
```

```python
df_counties = pd.read_csv('data/us-counties.csv', \
                               dtype={'fips': 'string'})
df_counties = df_counties[['state', 'fips']]
df_counties.dropna(inplace=True)
df_counties['fips_state'] = df_counties['fips'].apply(lambda x: x[:2])
```

```python
# join county level fips data to state data
df_states_copy = pd.merge(df_states, df_counties, how='left', on='fips_state')
df_states_copy = df_states_copy[['fips', 'cases']]
df_states_copy.dropna(inplace=True)
df_states_copy.shape
```

```python
from urllib.request import urlopen
import json
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)

import plotly.express as px

fig = px.choropleth_mapbox(df_states_copy, geojson=counties, locations='fips', color='cases_per_capita',
                           color_continuous_scale="Viridis",
                           range_color=(0, 0.07),
                           mapbox_style="carto-positron",
                           zoom=3, center = {"lat": 37.0902, "lon": -95.7129},
                           opacity=0.5,
                           labels={'cases_per_capita':'cases per capita'}
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
