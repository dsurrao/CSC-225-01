```python
import pandas as pd
import numpy as np
```

```python
cities = pd.read_csv('data/california_cities.csv')
```

```python
pd.set_option('display.max_rows', None)
cities.head()
```

```python
# Extract data we're interested in
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
```

```python
import matplotlib.pyplot as plt
```

```python
# Create a scatter plot
plt.scatter(lon, lat)
```

```python
# add size of points corresponding to city area
plt.scatter(lon, lat, s=area)
```

```python
# add a color map representing population (log10 of population)
# documentation: https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.scatter.html
plt.scatter(lon, lat, s=area, c=np.log10(population), cmap='viridis', alpha=0.5)
```

```python
# Add labels
plt.scatter(lon, lat, s=area, c=np.log10(population), cmap='viridis', alpha=0.5)
plt.axis(aspect='equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_10$(population)')
plt.clim(3, 7)
```
